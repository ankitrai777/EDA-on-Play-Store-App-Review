There are importantly three types of data : 
1.   First Type- that ends with 'M' represents size in 'Megabytes'
2.   Second Type- that ends with 'k' and represents size in 'kilobytes
3.   Third Type- "varies with device"

We wil convert these above mentioned types as below : 


1.   First Type - We will convert these data to float type as per kilobyte conversion value.
2.   Second Type - We will convert these data to float type.
3.   Third Type- Will convert these values to null values. After converting these values to null, we will fill them with catagorical median.

```python
# Checking for dtype of Size Column
ps_df['Size'].dtype
```

```python
# Changing Object dtype to float
def appropriating_size(x):
  if x == 'Varies with device':
    y= np.nan
  elif 'M' in x:
    m1= x.replace('M','')
    m2= float(m1)
    y= m2 * 1000
  elif 'k' in x :
    k1= x.replace('k','')
    y=float(k1)
  return (y)
```

```python
#Converting size in kb
ps_df['Size']=ps_df['Size'].apply(lambda a: appropriating_size(a))
```

# **3. HANDLING OUTLIERS**

Let's check out our 'Installs' distribution.

The graph is positively skewed and we have heavy Kurtosis , which mean lot of OUTLIERS !! For the necessity , we have to handle these outliers. Let's do it for all our columns other than object type.

```python
# Seggregating numeric columns from our ps_df

numeric_variable = [col for col in ps_df.columns if ps_df[col].dtype!='O']
num_df = ps_df[numeric_variable]
num_df.head()
```

```python
# Defining a function to check outliers

def check_outliers(dataframe, variable):
    low_limit, up_limit = outlier_limits(dataframe, variable)
    if dataframe[(dataframe[variable] < low_limit) | (dataframe[variable] > up_limit)].any(axis=None):
        print(variable, "yes")
    print(variable, "no")
```

```python
# Function for defining upper limit and lower limit
'This function will set up upper and lower limits for our data. Values outside this will be considered - Outliers.'

def outlier_limits(df1, col_name):
    quartile1 = df1[col_name].quantile(0.25)
    quartile3 = df1[col_name].quantile(0.75)
    interquantile_range = quartile3 - quartile1
    upper_limit = quartile3 + 1.5 * interquantile_range
    lower_limit = quartile1 - 1.5 * interquantile_range
    return lower_limit, upper_limit
```

```python
#Function to find and adjust outliers to their nearest limit
'This function will check for the outliers and replace them with their nearest limits.'

def adjusting_outliers(df2, numeric_columns):
    for variable in numeric_columns:
        lower_limit, upper_limit = outlier_limits(df2, variable)
        df2.loc[(df2[variable] < lower_limit), variable] = lower_limit
        df2.loc[(df2[variable] > upper_limit), variable] = upper_limit
```

```python
# Applyling above defined function to our numeric columns

adjusting_outliers(num_df, num_df.columns)
```

```python
# Checking if outliers are adjusted

for col in num_df.columns:
    check_outliers(num_df, col)
```

```python
# Plotting distribution of sentiment of apps among all playstore aps

app_plot = ur_df['Sentiment'].value_counts().plot (kind = 'pie',autopct='%1.1f%%')
app_plot.set_title("Distribution of Sentiments across apps")
```

Which are the top apps for which users registered their comments ?

```python
# Top Apps based on their number of comments

ur_df['App'].value_counts().sort_values(ascending = False).reset_index().head(5)
```

Going throgh just one type of sentiments - Positive or Negative, will leave us at any better position to analyse the sentiments. Let's define a new parameter which willl use these sentiments to showcase the success of an app.


**Ratio of Positive to Negative Comments :** If value of positive to negative ratio is < 1, then the overall sentimetns are negative and if value is >1, then the overall sentiment is positive. However, larger the value from 1 , it will show more positive sentiments for that App.

Let's get job done !!

```python
# Let's unstack sentiments and their numbers 
new = ur_df[['App','Sentiment']].groupby('App')['Sentiment'].value_counts()
new_df = new.unstack(level = -1)
new_df.head()
```

```python
# Giving life to a new parameter

new_df['Sentiment Ratio'] = round(new_df['Positive']/new_df['Negative'] ,0)
new_df.head()
```

Great !! WE HAVE A NEW PARAMETER TO HELP ANALYSING THE OVERALL SENTIMENT. What is the median ratio our apps are giving in terms of sentimer ??

```python
new_df['Sentiment Ratio'].median()
```

**CONCLUSION:** Nice!! Seems like most apps have four times more positive comments for their app compared to the negative comments.
